```python
#Cleaning Data:
#URLs,Hashtags,mentions,...
import re
def cleanResume(txt):
    cleanText = re.sub('http\S+\s', ' ', txt)
    cleanText = re.sub('RT|cc', ' ', cleanText)
    cleanText = re.sub('#\S+\s', ' ', cleanText)
    cleanText = re.sub('@\S+', '  ', cleanText)  
    cleanText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) 
    cleanText = re.sub('\s+', ' ', cleanText)
    return cleanText
cleanResume("The quick brown £$ fox jumps over the lazy#### dog http://helloword txt@gmail.com")
```

```python
df['Resume'] = df['Resume'].apply(lambda x:cleanResume(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.corpus import stopwords

# Télécharger les stop words de nltk (à faire une fois)
nltk.download('stopwords')

# Récupérer les stop words en anglais et en français
stop_words_english = stopwords.words('english')
stop_words_french = stopwords.words('french')

# Combiner les listes de stop words
stop_words_combined = list(set(stop_words_english + stop_words_french))

# Utiliser cette liste dans TfidfVectorizer
tfidf = TfidfVectorizer(stop_words=stop_words_combined)



# Adapter et transformer les CVs
tfidf.fit(df['Resume'])
requiredText = tfidf.transform(df['Resume'])

```

```python
"""Diviser les données en ensembles d'entraînement 
et de test est une étape cruciale dans 
le processus de construction et d'évaluation des modèles de machine learning"""

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(requiredText, df['Category'], test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
from docx import Document

# Fonction de prédiction pour un nouveau CV à partir d'un fichier .docx
def predict_resume_from_file(file_path):
    # Lire le fichier .docx et extraire le texte
    doc = Document(file_path)
    new_resume = '\n'.join([para.text for para in doc.paragraphs])
    
    # Nettoyer et transformer le texte
    cleaned_resume = cleanResume(new_resume)
    transformed_resume = tfidf.transform([cleaned_resume])
    
    # Prédiction
    predicted_category = clf.predict(transformed_resume)
    predicted_category_label = le.inverse_transform(predicted_category)
    return predicted_category_label[0]

def extract_years_experience_from_file(file_path):
    # Lire le fichier .docx et extraire le texte
    doc = Document(file_path)
    text = '\n'.join([para.text for para in doc.paragraphs])
    
    # Regex pour trouver des phrases comme "X years" ou "X ans"
    match = re.search(r'(\d+)\s*(ans|years?)', text, re.IGNORECASE)
    if match:
        return int(match.group(1))
    else:
        return 0  # Retourner 0 si aucune année d'expérience n'est trouvée
    
    
def The_Score(years_experience):
    if 1 < years_experience < 10:
        score = (years_experience / 10) * 100
        return f"{score:.2f}%"  # Utilisation de f-string pour formater le score avec deux décimales après la virgule
    else:
        return '0%'
    
Path='C:/Users/R I B/Desktop/Nouveau dossier/dataset/OPERAM - CV- HICHEM - IED PHP DRUPAL.docx'    
years_experience = extract_years_experience_from_file(Path)
predicted_label = predict_resume_from_file(Path)
print(f"The predicted category for the new resume is: {predicted_label}")
print(f"Years of experience extracted: {years_experience}")
Score=The_Score(years_experience)
print(f"Score: {Score}")

```
